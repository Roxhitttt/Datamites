# 1. Check for Even Numbers: Prints if a number is even using modulo.

```python
num = (1,3,45,2,8,9,4,7,0)
for i in num:
    if i % 2 == 0:
        print(i)
```

```python
for i in range(20):
    if i % 2 == 0:
        print(f"{i} is even")
```

# 2. Check for Odd Numbers: Prints if a number is odd using !=.

```python
num = (1,3,45,2,8,9,4,7,0)
for i in num:
    if i % 2 != 0:
        print(i)
```

```python
for i in range(20):
    if i % 2 != 0:
        print(f"{i} is odd")
```

# 3. Sum of Positives: Adds numbers only if they are greater than zero.

```python
num = (1,-3,45,2,-8,9,-4,7,0)
sumofallnum = 0
for i in num:
    if i > 0:
        sumofallnum += i
print(sumofallnum)
```

```python
for i in (1,-3,45,2,-8,9,-4,7,0):
    if i < 0:
        print(f"Negative num found {i}")
```

# 7. Floor Division Check: Checks if a number fits into another evenly.

floor division : is performed using the double slash operator (//). It divides two numbers and rounds the result down to the nearest whole integer, effectively discarding any decimal remainder.

* Standard Division ( / ) : 7 / 2 = 3.5

* Floor Division ( // ) : 7 // 2 = 3

* Example 1: Iterating Through a List

The most common way to use floor division with a for loop is to process a list of numbers (numerators) against a specific divisor.

```python
numbers = [10, 15, 23, 50, 9]
divisor = 4

print(f"Dividing by {divisor} using floor division (//):")
print("-" * 30)

for num in numbers:
    result = num // divisor
    print(f"{num} // {divisor} = {result}")
```

```python
for i in range(10, 20):
    if i // 5 == 3: # Intergers 15 to 19
        print(f"{i} divided by 5 is 3 (floored)") 
```

# 10. Absolute Value Check: Using abs() with comparison.

The abs() function in Python returns the absolute value of a number. In mathematics, this is written as | x |. It represents the distance of a number from zero, regardless of direction.





```python
deltas = [-10, 5, -1, 0]
for d in deltas:
    if abs(d) > 4:
        print(f"Large deviation: {d}")
```

```python
deltas = [-10, 5, -1, 0]
for d in deltas:
        abs(d)
        print(f"Large deviation: {d}")
```

```python
deltas = [-10, 5, -1, 0]
for d in deltas:
    if abs(d) > 0:
        print(f"Large deviation: {d}")
```

# Group B: Logical Operators (11-20)

Logical operators are used to combine multiple conditions into a single check. They evaluate expressions and return either True or False

* and	: Returns True only if both sides are true. 

* or : 	Returns True if at least one side is true.

* not :	Reverses the result (True becomes False).

Focus : and, or, not.

# 1. Range Check (AND): Number is between 10 and 20.

```python
for i in range(30):
    if i > 10 and i < 20:
        print(f"{i} is in the golden zone")
```

# 2. Either/Or Condition (OR): Number is extremely high or extremely low.

```python
temprature = [-5, 10, 25, 105]
for t in temprature:
    if t < 0 or t > 100:
        print(f"Extreme Weather : {t} ")
```

# 3. Not Logic (NOT): Check if a flag is False.

```python
flags = [True, False, True]
for f in flags:
    if not f:
        print("System is down")
```

# 4. Complex AND/OR: FizzBuzz Logic (Divisible by 3 and 5).

```python
for i in range(1, 16):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
```

# 6. String Length and Content: Two conditions on a string.

```python
words = ["apple", "to", "banana", "it"]
for w in words:
    if len(w) > 3 and "a" in w:
        print(w)
```

# 7. Multi-variable check: Iterating tuple unpacking.

```python
points = [(1, 2), (5, 5), (10, 1)]
for x, y in points:
    if x > 3 or y > 3:
        print(f"Coordinate {x},{y} is far out")
```

# 8. Not Equal Multiple: Excluding specific numbers.

```python
for i in range(5):
    if i != 2 and i != 4:
        print(i)
```

# 10. Combined Arithmetic and Logic:

```python
for i in range(10):
    if (i + 5) > 10 and (i - 2) < 5:
        print(i)
```
