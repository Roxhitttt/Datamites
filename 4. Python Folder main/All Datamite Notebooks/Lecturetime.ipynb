```python
# Creating a list 
numbers = [10, 34, 23, 54, 56,"Aditya","Rohit","Anybody"]

# Accessing elements 
print(numbers[2])

# Modifying elements 
numbers[1] = 87
print(numbers)

# Adding elements
numbers.append(11) # append is used to add new value to the list at the end
print(numbers)

# Adding elements in a specific location
numbers.insert(1,"Gaurav") # if we want to add elements in a specific location then we use insert function,

# removing elements
numbers.remove(11)
print(numbers)

# removing values using pop
numbers.pop() # bydefault last value will be removed

# removing values using pop in a specific location
numbers.pop(6)
```

# Difference between list and tuple 

The main difference between lists and tuples in programming, particularly in Python, is their mutability.

* Lists are mutable, meaning you can change their contents (add, remove, or modify elements) after they are created. They are typically defined using square brackets ([]).

* Tuples are immutable, meaning you cannot change their contents once they are created. They are typically defined using parentheses (()).

# Sets 
A Set in Python is a collection of unique elements that is unordered and mutable.

Think of it like a bag of distinct items where the order you put them in doesn't matter, and duplicates are automatically thrown out.

## Key Features
* Unique Elements: Sets do not allow duplicate values. If you try to add a duplicate, it is ignored.

* Unordered: Items have no fixed position. You cannot access items using an index (e.g., myset[0] will raise an error).

* Mutable: You can add or remove items after the set is created.

* Syntax: Defined using curly braces {} or the set() function.

```python
# Creating a set
fruits = {"apple", "banana", "orange"}
# Adding elements
fruits.add("grape")
print(fruits)
# Removing elements 
fruits.remove("banana")
print(fruits)
```

# Dictionaries

A Dictionary in Python is a collection that stores data in key-value pairs.

* Key-Value Structure: Every item consists of a pair: Key: Value.

* Fast Lookup: Designed to retrieve data very quickly using the key.

* Mutable: You can add, remove, or change items after creation.

* Unique Keys: Keys must be unique (no duplicates allowed) and immutable (e.g., strings, numbers). Values can be anything.

```python
full_name = "Rohit Gosavi"
print(full_name[3:10])
print(full_name[3:])
print(full_name[:9]) 

print(full_name[::2]) # Step size will skip 1 character
```

# Loop

a loop is a sequence of instructions that repeats itself until a specific condition is met.

There are 2 types of loop:

1. For loop - Use this when you know exactly how many times you want to repeat something.

3. While loop - Use this when you don't know how many times you need to repeat, but you know the condition for stopping.

* What is iterator : An iterator is an object that remembers its position in a sequence of data and delivers the next item whenever the loop asks for it.

```python
name = "Praful"
for c in name: # here c is iterater.
    print(c)
```

```python
name = "Praful"
for c in name[::-1]: # if we want to print name in reverse order
    print(c)
```

```python
name = "Praful"
for c in name[::2]: # if we want to skip 2 characters
    print(c)
```

# Print all the even numbers from 1 to 151 

* range (1,151,1)

It follows the pattern: range(start, stop, step)

Start: Where the count begins.

Stop: Where the count ends, meaning it stops before this number i.e it will stop at 150 cuz we gave 151

Step: # we use stepsize to skip 1 value in sequence

```python
for num in range(1,151):
    if num % 2 == 0:
        print(num)
```

```python
# skip 1 value in sequence

for num in range(2,100,2): # we use step size to skip 1 value in sequence
    print(num)
```
