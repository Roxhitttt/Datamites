```python
# Function Defination
def greet():
    print("Hello, There!")

# Calling a Function
greet()

# Function Parameters
def greet(name):
    print("Hello, " + name + "!")
greet("Alice")

# Function Return Value 
def multiply(a, b):
    product = a * b
    return product
result = multiply(5, 3)
print(result)

# Example
def calculate_area(length, width):
    area = length * width
    return area
length = 5
width = 3
result = calculate_area(length, width)
print(result)
```

```python
# Function Arguments : Python support different types of function arguments, including positional arguments, keyword arguments and default arguments.

# Positional Arguments : in the same order as the function defination

def greet(name, age):
    print("Hello, " + name + "! You are " + str(age) + " years old.")
greet("Alice", 25)

# Keyword arguments: regardless of their position

def greet(name, age):
    print("Hello, " + name + "! You are " + str(age) + " years old.")
greet(age=30, name="Bob")

# Default Arguments: defalult arguments are parameters that have a predefined value.

def greet(name, greeting="Hello"):
    print(greeting + ", " + name + "!")
greet("Alice")
greet("Bob", greeting="Hi")
```

```python
# Function can return multiple values by separating them with commas.

def calculate(a, b):
    sum = a + b
    difference = a - b
    return sum, difference
result_sum, result_difference = calculate(10, 5)
print(result_sum)
print(result_difference)
```

```python
# Example 

def calculate_rectangle_properties(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter
resultarea, resultperimeter = calculate_rectangle_properties(10, 5)
print(resultarea)
print(resultperimeter)
```

# Lambdas : Anonymous Functions

Lambdas, provide a concise way to define small, one-line functions without a formal function name, Useful in situations where a function is required for a short and specific purpose.

Syntax : 
lambda arguments: expression

lambda keyword is followed by a list of arguments which represets the inputs to the lambda.

```python
# Example of lambda function that calculates the square of a number:

square = lambda x: x ** 2
```

# Using Lambdas:

```python
# Sorting : Lambdas can be used as key functions in sorting operations to define custom sorting criteria

fruits = ['apple', 'banana', 'cherry', 'date']
sorted_fruits = sorted(fruits, key = lambda x: len(x)) # the lambda function 'lambda x: len(x)' is used as the key function to sort the list of fruits based on their lengths 
print(sorted_fruits)

# Filtering : Lambdas can be used in combination with built-in functions like 'filter()' to filter elements from a list based on specific conditions
#Example : filters even numbers from a list 

numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)

# Mapping : Lambdas can be used with functions like 'map()' to apply a specific operation to each element of a list,
#Example : Doubles each number in a list 

numbers = [1, 2, 3, 4, 5]
doubled_numbers = list(map(lambda x: x * 2, numbers))
print(doubled_numbers)
```

```python
# Example : Consider a scenario where you have a list of students, and you need to sort them based on their names.

students = [
    {'name': 'Alice', 'age': 20},
    {'name': 'Bob', 'age': 19},
    {'name': 'Charlie', 'age': 21}
]
sorted_students = sorted(students, key=lambda x: x['name'])
print(sorted_students)

# in this example, lambda function 'lambda x: x['name'] is used as the key function to sort the list of students based on their names
```

# Understanding Scope of Variables in Functions

The concept of variable scope defines the accessibility and lifespan of variables within a program

1. Local Scope : When a vairable is defined inside a function, it has local scope. This means that the variable is only accessible within the funtion where it is defiend.

```python
# Ex

def my_function():
    x = 10
    print(x)
my_function()
print(x) # Error : NameError: name 'x' is not defined

# Local variables are created when the function is called and destroyed when the function completes execution. They are not accessible outside the funtion
```

2. Global Scope : Variables defined outside any function or block have global scope, meaning they can be accessed from anywhere in the program.

```python
# Ex

x = 10
def my_function():
    print(x)
my_function()
print(x)

# Global vairables can be accessed and modified from any part of the program
```

```python
# However, modifying global variables within a function requires explicitly decalring them as global using the 'global' keyword.
```

```python
# Modifying Global Variables within a function
#Ex

x = 10
def modify_global():
    global x
    x = 20
modify_global()
print(x)

# It is generally considered good practice to minimize the use of global variables.
```

```python
# Example : Consider a scenario where you have a global variable 'total' representing the total sales of a company. You want to calculate the bonus
# amount for each  employee based on their individual sales, while keeping track of the overall sales total.

total = 0
def calculate_bonus(sales):
    global total
    bonus = sales * 0.1
    total += sales # total = total + sales
    return bonus
employee1_sales = 5000
employee1_bonus = calculate_bonus(employee1_sales)
print("Employee 1 Bonus:", employee1_bonus)
print("Total Sales:", total)
```
