```python
def greet(name):
    # Outer code block
    print("Hello, "+ name + "!")
    if name == "Alice":
        # Nested code block 
        print("You are amazing, " + name + "!")
    else:
        # Another nested code block
        print("Nice to meet you, " + name + "!")
    # Code outside the if-else block 
    print("Have a great day, " + name + "!")
# Function call
greet("Rohit")
```

# LIsts

```python
# Creating a list 
numbers = [10, 34, 23, 54, 56,"Aditya","Rohit","Anybody"]

# Accessing elements 
print(numbers[2])

# Modifying elements 
numbers[1] = 87
print(numbers)

# Adding elements
numbers.append(11) # append is used to add new value to the list at the end
print(numbers)

# Adding elements in a specific location
numbers.insert(1,"Gaurav") # if we want to add elements in a specific location then we use insert function,

# removing elements
numbers.remove(11)
print(numbers)

# removing values using pop
numbers.pop() # bydefault last value will be removed

# removing values using pop in a specific location
numbers.pop(6)
```

# Sets

```python
# Creating a set
fruits = {"apple", "banana", "orange"}
# Adding elements
fruits.add("grape")
print(fruits)
# Removing elements 
fruits.remove("banana")
print(fruits)
```

# Dictionary

```python
# Creating a dictionary
person = {
    "name" : "John Doe",
    "age" : 25, 
    "city" : "London"
}
# Accessing values
print(person["name"])
print(person.get("age"))
print(person["city"])
# Modifying values
person["age"] = 30
print(person)
```
